```python
def gain(
    isequence0, isequence1, xslice=slice(4, 2044), yslice=slice(4, 2044), verbose=True
):
    filepathlist = sorted(glob.glob(datadirectory + "*.fits"))[:]
    meanlist = []
    varlist = []
    for i in range(2,12):
        print(i)
        filepath0 = filepathlist[sequenceslice(isequence0)][i]
        filepath1 = filepathlist[sequenceslice(isequence1)][i]
        hdu0 = astropy.io.fits.open(filepath0)[0]
        hdu1 = astropy.io.fits.open(filepath1)[0]
        header0, data0 = hdu0.header, hdu0.data.astype(np.float32)
        header1, data1 = hdu1.header, hdu1.data.astype(np.float32)
        data0[np.where(data0 == 0)] = np.nan
        data1[np.where(data1 == 0)] = np.nan
        data0 = h2rg.channelcorrection(data0, order=1)
        data1 = h2rg.channelcorrection(data1, order=1)
        data = data1 - data0
        data = h2rg.channelcorrection(data, order=1)
        mean0 = np.nanmean(data0)
        mean1 = np.nanmean(data1)
        var = astropy.stats.mad_std(data[yslice, xslice], ignore_nan=True) ** 2
        if verbose:
            print(
                "%.2e %.2e %5.2f %s  %s" % (mean0, mean1, var, filepath0, filepath1)
            )
        mean = 0.5 * (mean0 + mean1)
        meanlist.append(mean)
        varlist.append(var)

    meanlist = np.array(meanlist)
    varlist = np.array(varlist)

    plt.plot(meanlist, varlist, linestyle="none", marker="+", label="Data")
    p = np.polyfit(meanlist, varlist, 1)
    print(p)
    plt.plot(meanlist, meanlist * p[0] + p[1], label="Model")
    plt.ylim(0, 20e3)
    plt.xlim(0, 25e3)

    plt.legend()
    plt.xlabel("Mean (DN)")
    plt.ylabel("Variance in Difference (DN$^2$)")
    plt.savefig("gain-%d-%d.png" % (isequence0, isequence1))
    plt.show()

    gain = 2 / p[0]
    print("gain = %.2f" % gain)

gain(0, 1)
gain(2, 3)
gain(4, 5)
gain(6, 7)
```
